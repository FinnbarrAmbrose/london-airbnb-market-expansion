### 1. Project Background

A property management company is exploring opportunities to expand its short-term rental operations within London.
Before committing capital to new properties, leadership requires data-driven insight into which areas and property types offer the strongest revenue potential while balancing pricing, demand, and operational quality.

### 2. Business Problem

Expanding into the short-term rental market carries financial risk if investments are made in areas with weak demand, inconsistent pricing, or poor host performance.

The business needs evidence-based guidance to:

Identify high-potential neighbourhoods

Understand pricing dynamics

Assess market maturity and quality indicators

### 3. Business Question

Which London neighbourhoods and property types offer the best revenue potential for a short-term rental expansion, based on pricing, demand, and host performance?

This question is:

Decision-focused

Measurable using public data

Directly actionable for market expansion strategy

### 4. Success Metrics (KPIs)

The following metrics will be used to evaluate performance across neighbourhoods and property types:

#### 1️⃣ Median Nightly Price (£)

#### Reduces distortion from luxury outliers

Enables fair comparison between areas

#### 2️⃣ Estimated Occupancy Proxy (%)

Derived from review activity and availability

Used as a proxy for demand where true occupancy is unavailable

#### 3️⃣ Estimated Annual Revenue (£)

Combines price and demand into a single business metric

Supports ranking of investment opportunities

#### 4️⃣ Superhost Share (%)

Proxy for service quality and operational maturity

Helps identify stable vs higher-risk markets

### 5. Assumptions & Constraints

To ensure transparency, the following assumptions apply:

True occupancy rates are not directly available

Review counts are used as a proxy for demand

Revenue estimates are directional, not guarantees

Analysis supports strategic decisions, not financial forecasts

These assumptions will be revisited and validated where possible during analysis.
